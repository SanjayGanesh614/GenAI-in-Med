```python
import torch
from utils import evaluate


def squeeze_features(X, bit_depth=4):
    X = X.clone()
    X = ((X * (2**bit_depth - 1)).round() / (2**bit_depth - 1)).float()
    return X

def evaluate_with_feature_squeezing(model, X_test, y_test, scaler):
    X_tensor = torch.tensor(X_test).float()
    y_tensor = torch.tensor(y_test).long()
    squeezed_X = squeeze_features(X_tensor)

    acc_orig = evaluate(model, X_tensor, y_tensor)
    acc_squeezed = evaluate(model, squeezed_X, y_tensor)

    print(f"Accuracy on original inputs: {acc_orig*100:.2f}%")
    print(f"Accuracy on squeezed inputs: {acc_squeezed*100:.2f}%")
```
